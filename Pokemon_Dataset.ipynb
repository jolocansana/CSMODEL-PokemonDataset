## 7. Image Processing on the Pokemon Dataset

In this section, we answer the research question:

> Is there a relationship between the dominant color of the Pokemon sprite and a certain type?

In order to find correlations on the dataset, let's find the dominant colors. In the following cells, we'll use Pikachu to show an example of the algorithm we are going to use for the entire dataset to find the dominant colors per image. First, we import the require libraries:
- matplotlib.image: to load images as a scatter plot of pixels
- matplotlib.pyplot: to render the plot
- scipy.cluster.vq.whiten: to normalize the pixel values
- scipy.cluster.vq.kmeans2: to perform kmeans clustering

Now, lets import the pikachu image from our image folder.

````
import matplotlib.image as img 
import matplotlib.pyplot as plt 
from scipy.cluster.vq import whiten 
from scipy.cluster.vq import kmeans2 

pikachu = matplotlib.image.imread(dirname + 'pikachu.png')
plt.imshow(pikachu)
````

It should be noted that the images in the dataset contain a transparent background, and that matplotlib.image stores the rgba values of each image where the transparent pixels have an A value of 0. We can thus eliminate the transparent pixels if the alpha values are equal to 0 to disregard the background in clustering. To start the clustering process, let us preprocess the data into something the kmeans function can use. We get the RGB values from each pixel in the matplotlib.image 'pikachu' and append them to their own arrays. From these arrays, we then create a dataframe that will start the RGB values of the image.

Note that we set our k parameter for kmeans as 3. Due to manual testing, we found that clusters of k=1 and k=2 caused the dominant color to be too mixed with other colors to find a distinguished dominant color, while 4 clusters began to introduce dominant colors that were mostly insignficant. Thus, we set our k as 3. In order to determine the most dominant color of the three clusters, by retrieving the bin count of each cluster, we simply choose the cluster with the highest bin count. 

````
r = [] 
g = [] 
b = [] 

# multiply colors by 255, as rgb values in matplot.image saves pixels as rgb values scaled from 0 -> 1
for row in pikachu: 
    for temp_r, temp_g, temp_b, temp_alpha in row:
      if (temp_alpha != 0):
        r.append(temp_r * 255) 
        g.append(temp_g * 255) 
        b.append(temp_b * 255) 
   
image_df = pd.DataFrame({'red' : r, 
                          'green' : g, 
                          'blue' : b}) 
````

Then, we normalize the RGB values and attach it back to the dataframe so that we can cluster the colors properly.

````
# we whiten first before executing kmeans in order to normalize the data
image_df['scaled_color_red'] = whiten(image_df['red']) 
image_df['scaled_color_blue'] = whiten(image_df['blue']) 
image_df['scaled_color_green'] = whiten(image_df['green']) 
````

After normalization, we can now perform kmeans. We get kmeans by ploting the the scaled values to a scatter plot and define the number of clusters we want - in this case, 3. We get 3 because it seem to be the most accurate after some trial and errors. With a k of less than 3, we got disstorted colors, and with a k greater than 3, we got additional colors but the dominant color stayed the same. After getting the clusters, we then got the standard deviation of RGB respectively which we will use to normalize the centroids of the clusters generated from kmeans. That will yield the 3 dominant colors of the image. 

````
# execute kmeans with 3
# cluster_centers holds the k number of centroids
# clusters holds the cluster sets
cluster_centers, clusters = kmeans2(image_df[['scaled_color_red', 
                                    'scaled_color_green', 
                                    'scaled_color_blue']], 3) 
  
dominant_colors = [] 

# get standard deviation of each color
red_std, green_std, blue_std = image_df[['red', 
                                          'green', 
                                          'blue']].std() 

# to generate dominant colors, find cluster cetners, and scale with regards to standard dev
for cluster_center in cluster_centers: 
    red_scaled, green_scaled, blue_scaled = cluster_center 
    dominant_colors.append(( 
        red_scaled * red_std / 255, 
        green_scaled * green_std / 255, 
        blue_scaled * blue_std / 255
    )) 
  
plt.imshow([dominant_colors])
plt.show()
````

Since we want to most dominant color, from the 3 dominant, we get the max mode of 3 and return that color value.

````
# dominant = pd.Series(dominant_colors, np.bincount(clusters))
dominant = dominant_colors[np.where(np.bincount(clusters) == np.bincount(clusters).max())[0][0]]

plt.imshow([[dominant]])
plt.show()
plt.imshow(pikachu)

#   ALGORITHM ADAPTED FROM
#   https://www.geeksforgeeks.org/extract-dominant-colors-of-an-image-using-python/
````

In this case, we can see that Pikachu's most dominant color is this shade of yellow, and by visually expecting the actual image of Pikachu, we can see that the dominant color is correct.

Now that we now that the algorithm works, lets perform this to every Pokemon that has a image file.

## 8. Checking if Pokemon from different generations are statistically different in terms of stats using ANOVA and P-Tukey Analysis.

In this section, we answer the research question:

> Are Pokemon from different generations statistically different in terms of base stats?

To find that out, we will use ANOVA testing to check if there is difference statistically between a generation of pokemon to another. We will also use P-Tukey Analysis to check the significance from the ANOVA and perform post-hoc operations to identify the groups that are statistically different if the ANOVA is below our set confidence interval. First, we will set the confidence interval to 95%. We will set it to 95-percent because it is a common standard among different domains which means the P-value equals to 0.05. Therefore, if our P-value from the ANOVA testing is less than this amount, we will consider the set as statistically significant, in this case there is a difference between generations in that specific variable. To perform ANOVA, we will need an equal sample size per generation. We should check the total number of Pokemon per generation to find the most optimal sample size. 

````
pokemon_df.groupby('generation')['name'].count()
````

From the series above, we can see that the minimum number of Pokemon per generation is 72 of generation 6. Since we want to optimize the ANOVA tests, we should maximize the sample size n. Therefore, we are going to n = 72. Next, we should make data frames for each of the generations with their Pokemon's respective name, attack, defence, sp_attack, sp_defence, and speed base statistic. We will also include the base total which is the summation of all base stats and base hp. These statistics were selected to identify trends with the combat statistics and measure the Pokemon's overall strength.

````
import scipy.stats as stats
# stats f_oneway functions takes the groups as input and returns F and P-value
n = 72

generation_1 = pokemon_df[pokemon_df['generation'] == 1].sample(n)[['name', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed', 'base_total']]
generation_2 = pokemon_df[pokemon_df['generation'] == 2].sample(n)[['name', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed', 'base_total']]
generation_3 = pokemon_df[pokemon_df['generation'] == 3].sample(n)[['name', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed', 'base_total']]
generation_4 = pokemon_df[pokemon_df['generation'] == 4].sample(n)[['name', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed', 'base_total']]
generation_5 = pokemon_df[pokemon_df['generation'] == 5].sample(n)[['name', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed', 'base_total']]
generation_6 = pokemon_df[pokemon_df['generation'] == 6].sample(n)[['name', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed', 'base_total']]
generation_7 = pokemon_df[pokemon_df['generation'] == 7].sample(n)[['name', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed', 'base_total']]


generations = [generation_1, generation_2, generation_3, generation_4, generation_5, generation_6, generation_7]
````

Now let us check the first statistic 'attack' if there is any signficant difference between generations.

````
attack_gen = pd.DataFrame()

# make dataframe of attack of every gen
for x in range(0,7):
    attack_gen[x+1] = generations[x]['attack'].values

import statsmodels.api as sm
from statsmodels.formula.api import ols

# reshape the d dataframe suitable for statsmodels package 
d_melt = pd.melt(attack_gen.reset_index(), id_vars=['index'], value_vars=[1,2,3,4,5,6,7])
# replace column names
d_melt.columns = ['index', 'treatments', 'value']
# Ordinary Least Squares (OLS) model
model = ols('value ~ C(treatments)', data=d_melt).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
````

We did multiple ANOVA tests on different random samples per generations and have found that most PR(>F) or the P-value result from the ANOVA is above our set p-value of 0.05 therefore, the ANOVA test is not significant. We can continue to check this by performing pairwise tukey tests. This is an Ad Hoc test to check which specific group in the ANOVA test is statistically different but we can also use it to check between the relationship of two groups.

````
from pingouin import pairwise_tukey

m_comp = pairwise_tukey(data=d_melt, dv='value', between='treatments')
m_comp
````

From the Tukey table, we can see that in the p-tukey column, all of the values are greater than our prescribe p-value of 0.05. This further strengthens the conclusion that there is no difference between generations in terms the attack statistic that is statistically significant. Let us try to check the other 4 statistics if they are significant through ANOVA.

````
ANOVA_df = pd.DataFrame()
results = []

for stats in generations[0].columns[2:]:
    for x in range(0,7):
        ANOVA_df[x+1] = generations[x][stats].values

    # reshape the d dataframe suitable for statsmodels package 
    d_melt = pd.melt(ANOVA_df.reset_index(), id_vars=['index'], value_vars=[1,2,3,4,5,6,7])
    # replace column names
    d_melt.columns = ['index', 'treatments', 'value']
    # Ordinary Least Squares (OLS) model
    model = ols('value ~ C(treatments)', data=d_melt).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    results += [[stats, anova_table['PR(>F)'][0]]]

pd.DataFrame(results).rename(columns={0:'type', 1:'p-value'})
````

From the values above, we can conclude that there is no difference between generations across all statistics that is statistically significant. This may be because of the difference each types have on each other as tested using EDA when we found the most dominant type per base statistic in **section 6.A1**. By also looking at the table of means of each statistic of the pokemons per type at section **6.A2**, we can see that the type must have a strong influence on the ANOVA test we have made which may have made it result as such.
